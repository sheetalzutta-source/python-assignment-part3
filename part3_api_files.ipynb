```python

# Task 1 — File Read & Write Basics

# Part A — Write:
# Create a Python script that writes the following student notes to a file called python_notes.txt:

# Topic 1: Variables store data. Python is dynamically typed.
# Topic 2: Lists are ordered and mutable.
# Topic 3: Dictionaries store key-value pairs.
# Topic 4: Loops automate repetitive tasks.
# Topic 5: Exception handling prevents crashes.

# A.1 Write the above five lines to the file using write mode ('w'). Remember to include encoding="utf-8".
with open("python_notes.txt", "w", encoding="utf-8") as f:
    f.write("Topic 1: Variables store data. Python is dynamically typed.\n")
    f.write("Topic 2: Lists are ordered and mutable.\n")
    f.write("Topic 3: Dictionaries store key-value pairs.\n")
    f.write("Topic 4: Loops automate repetitive tasks.\n")
    f.write("Topic 5: Exception handling prevents crashes.\n")
print("File written successfully.")
print ()

# A.2. Append two more lines of your own to the same file using append mode ('a').
with open("python_notes.txt", "a", encoding="utf-8") as f:
    f.write("Topic 6: Functions encapsulate reusable code.\n")
    f.write("Topic 7: Modules organize code into files.\n")
print("Lines appended.")
print ()

# A.3. Print a confirmation message after each file operation (e.g., "File written successfully.", "Lines appended.").
print("Confirmation: File operations completed.")
print ()

# Part B — Read:
# B.1. Read the file back and print each line numbered, stripping the trailing newline character (\n) from each line

with open("python_notes.txt", "r", encoding="utf-8") as f:
    lines = f.readlines()

for i, line in enumerate(lines, start=1):
    print(f"{i}. {line.strip()}")

# B.2. Count and print the total number of lines in the file.
total_lines = len(lines)
print(f"Total number of lines: {total_lines}")
print ()

# B.3. Ask the user for a keyword and print all lines that contain that keyword (case-insensitive). 
# If no match is found, print a helpful message.
keyword = input("Enter a keyword to search for: ")
matches = [line for line in lines if keyword.lower() in line.lower()]

if matches:
    print(f"Lines containing '{keyword}':")
    print ()
    for line in matches:
        print(f"  {line.strip()}")
else:
    print(f"No lines found containing '{keyword}'.")
    print ()
```

```python
# Task 2 — API Integration

# Step 1 — Fetch and Display Products:
# 1.1. Make a GET request to fetch 20 products:
import requests

response = requests.get("https://dummyjson.com/products?limit=20")
products = response.json()

# 1.2. Parse the JSON response. Each product has fields including id, title, category, price, and rating.
parsed_products = products["products"]


# 1.3. Print a formatted table:

print(f"{'ID':<4} | {'Title':<45} | {'Category':<15} | {'Price':<9} | Rating")
print(f"{'-'*4}-|-{'-'*45}-|-{'-'*15}-|-{'-'*9}-|-------")
for product in parsed_products:
    print(f"{product['id']:<4} | {product['title']:<45} | {product['category']:<15} | ${product['price']:<8.2f} | {product['rating']}")

# Step 2 - Filter and Sort:
# 2.1. From the 20 fetched products, filter only those with a rating ≥ 4.5.
highly_rated = [p for p in parsed_products if p["rating"] >= 4.5]

# 2.2. Sort the filtered list by price in descending order.
sorted_products = sorted(highly_rated, key=lambda x: x["price"], reverse=True)

# 2.3. Print the filtered and sorted list.
print("\nHighly Rated Products (Rating ≥ 4.5), Sorted by Price (Desc):")
print(f"{'ID':<4} | {'Title':<45} | {'Category':<15} | {'Price':<9} | Rating")
print(f"{'-'*4}-|-{'-'*45}-|-{'-'*15}-|-{'-'*9}-|-------")
for product in sorted_products:
    print(f"{product['id']:<4} | {product['title']:<45} | {product['category']:<15} | ${product['price']:<8.2f} | {product['rating']}")

# Step 3 - Search by Category:
# 3.1. Make a GET request to fetch all products in the laptops category:
response = requests.get("https://dummyjson.com/products/category/laptops")
laptop_products = response.json()

# 3.2. Print the name and price of each laptop found.
print("\nLaptops:")
for laptop in laptop_products["products"]:
    print(f"{laptop['title']}: ${laptop['price']:.2f}")

# Step 4 - POST Request (Simulated):
# 4.1. Send a POST request to https://dummyjson.com/products/add with the mentioned JSON body:
new_product = {
    "title": "My Custom Product",
    "price": 999,
    "category": "electronics",
    "description": "A product I created via API"
}
response = requests.post("https://dummyjson.com/products/add", json=new_product)
created_product = response.json()

# 4.2. Print the full response returned by the server.

print(
    "\nNote: DummyJSON is a mock/test API. "
    "The POST endpoint simulates a create operation and returns a response with an assigned ID, "
    "but no data is actually stored on the server. "
    "This is standard behaviour for test APIs."
)
print(response.json())


```

```python
# Task 3 - Exception Handling

# Part A - Guarded Calculator:
# A.1 Write a function safe_divide(a, b) that performs division and handles exceptions:

def safe_divide(a, b):
    try:
        return a / b
    except ZeroDivisionError:
        return "Error: Cannot divide by zero"
    except TypeError:
        return "Error: Invalid input types"
    
# A.2. Test it with: safe_divide(10, 2), safe_divide(10, 0), safe_divide("ten", 2). Print each result.

print(safe_divide(10, 2))  # Should print 5.0
print(safe_divide(10, 0))  # Should print "Error: Cannot divide by zero"
print(safe_divide("ten", 2))  # Should print "Error: Invalid input types"

# Part B - Guarded File Reader:

# B.1. Write a function read_file_safe(filename) that handles file reading with exception handling:

def read_file_safe(filename):
    try:
        with open(filename, "r", encoding="utf-8") as f:
            content = f.read()
            print (f"Content of '{filename}':\n{content}")
    except FileNotFoundError:
        print(f"Error: File '{filename}' not found.")
    finally:
        print("File operation attempt complete.")

# B.2. Test it with "python_notes.txt" (should succeed) and "ghost_file.txt" (should fail gracefully).

print ()
read_file_safe("python_notes.txt")  # Should print the content of the file
read_file_safe("ghost_file.txt")  # Should print an error message

# Part C - Robust API Calls:
# C.1. Update your API code from Task 2 so that every requests.get() and requests.post() call is wrapped in a try-except block that handles per instructed:

import requests
def safe_get(url):
    try:
        response = requests.get(url, timeout=5)
        response.raise_for_status()  # Raises HTTPError for bad responses (4xx, 5xx)
        return response.json()
    except requests.exceptions.ConnectionError:
        print("Connection failed. Please check your internet.")
    except requests.exceptions.Timeout:
        print("Request timed out. Try again later.")
    except Exception as e:
        print(f"An unexpected error occurred: {e}")

# C.2. To use Timeout, pass a timeout parameter to your requests:
# requests.get(url, timeout=5) — this will raise Timeout if the server takes more than 5 seconds to respond.
def safe_post(url, data):
    try:
        response = requests.post(url, json=data, timeout=5)
        response.raise_for_status()  # Raises HTTPError for bad responses (4xx, 5xx)
        return response.json()
    except requests.exceptions.ConnectionError:
        print("Connection failed. Please check your internet.")
    except requests.exceptions.Timeout:
        print("Request timed out. Try again later.")
    except Exception as e:
        print(f"An unexpected error occurred: {e}")

# Part D — Input Validation Loop:
# Write a loop that:

# D.1. Asks the user: "Enter a product ID to look up (1–100), or 'quit' to exit: "
while True:
    user_input = input("Enter a product ID to look up (1–100), or 'quit' to exit: ")
    if user_input.lower() == "quit":
        print("Exiting the program. Goodbye!")
        break
    try:
        product_id = int(user_input)
        if 1 <= product_id <= 100:
            response = requests.get(f"https://dummyjson.com/products/{product_id}", timeout=5)
            if response.status_code == 404:
                print("Product not found.")
            elif response.status_code == 200:
                product_data = response.json()
                print(f"Product Title: {product_data['title']}, Price: ${product_data['price']:.2f}")
        
        else:
            print("Invalid input. Please enter an integer between 1 and 100.")
    
    except ValueError:
        print("Invalid input. Please enter a valid integer or 'quit'.")
    except requests.exceptions.ConnectionError:
        print("Connection failed. Please check your internet.")
    except requests.exceptions.Timeout:
        print("Request timed out. Try again later.")
    except Exception as e:
        print(f"An unexpected error occurred: {e}")
    
```
